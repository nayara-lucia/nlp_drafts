```python
import nltk
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
import glob
nltk.download("punkt") # The "punkt" package is a sentence and word tokenizer. It is used to split a text into smaller units like words or sentences, which is a fundamental step in many NLP tasks.
nltk.download("stopwords")

from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer
from sentence_transformers import SentenceTransformer
```

```python
def merge_docs():
    docs = []
    path = "/content/*.txt"
    files = glob.glob(f"{path}")
    for file in files:
        r = open(f"{file}","r",encoding="UTF-8")
        docs.append(r.read())
    return docs
```

```python
docs = merge_docs()
```

```python
print(docs[0])
```

```python
print(len(docs))
```

```python
sentences = []
for doc in docs:
    sentences.extend(sent_tokenize(doc))
```

## **Model**

```python
model = BERTopic(language="portuguese",
                 verbose=True,
                 top_n_words=15,
                 min_topic_size=10,
                 nr_topics=20,
                 vectorizer_model=CountVectorizer(ngram_range=(1,3),stop_words=stopwords.words("portuguese"), min_df=10))
```

```python
topic, prob = model.fit_transform(sentences)
```
